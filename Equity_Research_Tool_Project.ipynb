```python
from langchain_community.document_loaders import UnstructuredURLLoader
```

```python
loader = UnstructuredURLLoader(urls=[
    "https://en.wikipedia.org/wiki/Jallianwala_Bagh_massacre",
    "https://en.wikipedia.org/wiki/Indian_Rebellion_of_1857"
])

data = loader.load()
```

```python
print(data)
```

```python
!pip install langchain-text-splitters
```

```python
from langchain_text_splitters import CharacterTextSplitter
```

```python
splitter = CharacterTextSplitter(chunk_size=200, chunk_overlap=0,separator='\n')
# Step 1: Splitting will be done on the basis of the separator
# Step 2: Text will be merged if the chunk size is less than 200
```

```python
output = splitter.split_text(data[0].page_content)
```

```python
print(output)
```

```python
len(output)
```

```python
for chunk in output:
  print(len(chunk))
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
r_splitter = RecursiveCharacterTextSplitter(chunk_size=200, chunk_overlap=0, separators=['\n\n', '\n', '.', ' '])
chunks = r_splitter.split_text(data[0].page_content)
```

```python
for chunk in chunks:
  print(len(chunk))
```

```python
print(len(chunks))
```

```python
import pandas as pd
pd.set_option('display.max_colwidth', 100)
```

```python
df = pd.read_csv('data_spreadsheet.csv')
```

```python
df
```

```python
from sentence_transformers import SentenceTransformer
encoder = SentenceTransformer('all-mpnet-base-v2')
```

```python
vectors = encoder.encode(df.text)
```

```python
print(vectors)
```

```python
vectors.shape
```

```python
dim = vectors.shape[1]
```

```python
dim
```

```python
import faiss
index = faiss.IndexFlatL2(dim)
index
```

```python
index.add(vectors)
```

```python
search_query = encoder.encode('I will not eat fast food.')
print(search_query)
```

```python
import numpy as np
svec = np.array(search_query).reshape(1,-1)
svec.shape
```

```python
index.search(svec,k=2)
```
